## The mean of area

```python
licensees['Premise Zip Code'].value_counts().mean()
```

```python
X = licensees['Premise Zip Code'].value_counts()
```

## H0 : μ0 = 2.75

## H1 : μ0 # 2.75




```python
from scipy import stats



# Tính t-statistic và p-value
t_statistic, p_value = stats.ttest_1samp(X, 2.75)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```

### --> It's two slide
### Choose α = 0.1 -->: α/2 = 0.05

### p_value > α/2 

### Can't reject the hypothesis

### We can conclude that there are about 2 to 3 gun stores in an area in the US.
